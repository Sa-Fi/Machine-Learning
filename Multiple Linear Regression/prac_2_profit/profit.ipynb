```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# import csv file 
df = pd.read_csv('online.csv')
df.shape
```

```python
df.isnull().sum()
```

```python
x = df.drop(['Profit'],axis=1)
# x
y = df['Profit']
# y
```

```python
city = pd.get_dummies(x['Area'],drop_first=True)
# city
```

```python
x = x.drop('Area',axis=1) # jeheto area er value gula numerically convert korci so ei ketre ,area col ta , data set theke drop korte hobe
```

```python
x = pd.concat([x,city],axis=1)
x
```

```python
xtrain , xtest,ytrain,ytest = train_test_split(x,y,test_size=0.25)
```

```python
ytest
```

```python
reg = LinearRegression()
reg.fit(xtrain,ytrain)
reg.predict(xtest)
```

# R2 value for check the accuracy

```python
print("Accuracy = ",reg.score(xtest,ytest)*100)
```

```python
from sklearn.metrics import r2_score #using library function
score = r2_score(ytest,reg.predict(xtest))
score
```
