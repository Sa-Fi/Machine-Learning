```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

ds = pd.read_csv('car driving risk analysis.csv')

ds.head()

ds.shape

x = ds[['speed']]
y = ds['risk']

plt.scatter(ds['speed'],ds['risk'],marker='X',color='blue')
plt.xlabel('Speed at KM')
plt.ylabel('Risk')
plt.title('Risk at Speed')

from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest = train_test_split(x.values,y,test_size=0.2,random_state=1)

from sklearn.linear_model import LinearRegression

regression = LinearRegression()

regression.fit(xtrain,ytrain)
regression.predict(xtest)
plt.plot(xtest,ytest,color="red")
```
